```python
import pandas as pd
import requests
import config
import json
import numpy as np
```

```python
df = pd.read_csv('resources/csv/school_data.csv')
df.head()
```

```python
raw_data = []
for x in df['school_name'].tolist():
    url = f"https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={x}%2Cschool&inputtype=textquery&locationbias=circle%3A2000%4041.881832%2C-87.623177&fields=formatted_address%2Cname%2Crating%2Copening_hours%2Cgeometry&key={config.GOOGLE_API_KEY}"
    payload={}
    headers = {}

    response = requests.request("GET", url, headers=headers, data=payload).json()

    raw_data.append(response)
```

```python
raw_data
```

```python
lat = []
lng = []
form_addy = []
actual_name = []
candidates = []
for x in raw_data:
    try:
        lat.append(x['candidates'][0]['geometry']['location']['lat'])
        lng.append(x['candidates'][0]['geometry']['location']['lng'])
        form_addy.append(x['candidates'][0]['formatted_address'])
        actual_name.append(x['candidates'][0]['name'])

    except:
        lat.append(np.NaN)
        lng.append(np.NaN)
        form_addy.append(np.NaN)
        actual_name.append(np.NaN)
    
df['lat'] = lat
df['lng'] = lng
df['form_addy'] = form_addy
df['actual_name'] = actual_name
df
```

```python
df.to_csv('resources/csv/googled_school_data.csv', index=False)
```
